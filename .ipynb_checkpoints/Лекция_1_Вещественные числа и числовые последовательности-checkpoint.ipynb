## Основные аксиомы теории чисел

Теория чисел, как математическая дисциплина, опирается на несколько фундаментальных аксиом. Эти аксиомы задают базовые правила и свойства чисел, на которых строится вся теория.

__1. Аксиомы для натуральных чисел:__
* 1 – натуральное число. (Существует "начало" счетного ряда).
* Каждое натуральное число имеет преемник, который также является натуральным числом. (Мы можем считать дальше).
* 1 не является преемником ни одного натурального числа. (Единица - это "первое" натуральное число).
* Разные натуральные числа имеют разные преемники.
* Принцип математической индукции: Если какое-то свойство верно для 1 и если из предположения, что оно верно для натурального числа n, следует его верность для n+1, то это свойство верно для всех натуральных чисел. (Это позволяет нам доказывать утверждения о всех* натуральных числах).

__2. Аксиомы арифметики:__
* Ассоциативность сложения и умножения: (a + b) + c = a + (b + c), (a ⋅ b) ⋅ c = a ⋅ (b ⋅ c)
* Коммутативность сложения и умножения: a + b = b + a, a ⋅ b = b ⋅ a
* Дистрибутивность: a ⋅ (b + c) = a ⋅ b + a ⋅ c

### Свойства рациональных чисел

#### Счетность рациональных чисел:

Что значит "счетное множество"?


Множество называется счетным, если его элементы можно поставить в взаимно однозначное соответствие с натуральными числами (1, 2, 3,...). Другими словами, мы можем пронумеровать все элементы этого множества.

__Теорема__ 

Множество рациональных чисел счетно.

__Доказательство:__


1. Представим рациональные числа как дроби p/q, где p и q — целые числа.
* Обратите внимание: мы рассматриваем только несократимые дроби (p и q взаимно просты).

2. Создадим бесконечную таблицу, в которой строки представляют числители (p), а столбцы - знаменатели (q):


     1  2  3  4 ...

   1 1/1 1/2 1/3 1/4 ...

   2 2/1 2/2 2/3 2/4 ...

   3 3/1 3/2 3/3 3/4 ...

   ... ... ... ... ...





4. Перебираем элементы таблицы по диагоналям:


* Первая диагональ: 1/1, 2/1, 3/1,...
* Вторая диагональ: 1/2, 2/2, 3/2,...

4. Пронумеруем все рациональные числа по порядку их появления в этой последовательности.

__Вывод:__

Мы построили взаимно однозначное соответствие между множеством всех рациональных чисел и множеством натуральных чисел. Это доказывает, что множество рациональных чисел счетное.

#### Плотность множества рациональных чисел


__1. Предположение:__ Пусть a и b —  произвольные рациональные числа, где a < b.

__2. Определение:__ По определению рационального числа, a = p/q и b = r/s, где p, q, r, s - целые числа, а q и s не равны нулю.


__3. Среднее значение: Рассмотрим среднее значение a и b:__

c = (a + b) / 2 = (p/q + r/s) / 2 = (ps + qr) / (2qs).

__4. Рациональность c:__
* ps + qr — целое число (сумма целых чисел).
* 2qs - целое число (произведение целых чисел, q и s не равны нулю).


Следовательно, c = (ps + qr) / (2qs) является рациональным числом.

__5. Неравенство:__ a < c < b:
* a < c: поскольку a < b, то (a + b)/2 > a
* c < b: поскольку a < b, то (a + b)/2 < b


__Вывод:__



Мы показали, что для любых двух рациональных чисел a и b (где a < b) существует рациональное число c такое, что a < c < b. Это доказывает утверждение.

## Числовые последовательности

### Определение и примеры числовых последовательностей

Числовая последовательность - это особый вид последовательности, где все элементы являются числами.

Проще говоря, это просто список чисел, записанных в определенном порядке.

Примеры числовых последовательностей:

* 1, 2, 3, 4, 5... (последовательность натуральных чисел)
* 2, 4, 6, 8, 10... (последовательность четных чисел)
* 1, 4, 9, 16, 25... (последовательность квадратов натуральных чисел)

Числовые последовательности широко используются в математике для описания различных закономерностей и зависимостей.

1. Арифметическая прогрессия:

* $a_n = a_1 + (n-1)d$
* $a_n$ - n-ый член последовательности
* $a_1$ - первый член последовательности
* $d$ - разность арифметической прогрессии

2. Геометрическая прогрессия:

* $a_n = a_1 \cdot q^{n-1}$
* $a_n$ - n-ый член последовательности
* $a_1$ - первый член последовательности
* $q$ - знаменатель геометрической прогрессии

3. Последовательность Фибоначчи:

* $F_1 = 1$, $F_2 = 1$, $F_n = F_{n-1} + F_{n-2}$ для $n > 2$


-----


__Определение верхней грани последовательности:__

Число M называется __верхней гранью__ числовой последовательности $(a_n)$, если для любого элемента $a_n$ этой последовательности выполняется неравенство:


$$ a_n \le M $$


Другими словами: __Верхняя грань__ - это число, которое больше или равно всех членов последовательности.


* Верхняя грань последовательности $\left(\frac{1}{n}\right)$ равна 1, то есть: $$\sup_{n\in\mathbb{N}}\frac{1}{n}=1.$$
* Верхняя грань последовательности $\left(\frac{n-1}{n}\right)$:   $$\sup_{n\in\mathbb{N}}\left(\frac{n-1}{n}\right)=1.$$


Число L называется __нижней гранью__ числовой последовательности $(a_n)$, если для любого элемента $a_n$ этой последовательности выполняется неравенство:


$$ a_n \ge L $$



* Нижняя грань последовательности $\left(\frac{1}{n}\right)$: $$\inf_{n\in\mathbb{N}}\frac{1}{n}=0.$$
* Нижняя грань последовательности $\left(\frac{n-1}{n}\right)$:   $$\inf_{n\in\mathbb{N}}\left(\frac{n-1}{n}\right)=0.$$



__Важно:__


* Последовательность может иметь несколько верхних (нижних) граней.
* Если последовательность не ограничена сверху (снизу), то она не имеет верхней  (нижней) грани.

-----


```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Генерируем значения n для последовательностей
n = np.arange(1, 50, 1)  

# Вычисляем члены последовательностей
y1 = math.e ** (-0.5*n)

plt.plot(n, y1, marker='o', linestyle='-', color='b', label='1/n')  

plt.show() 
```


### Фундаментальная последовательность

__Определение фундаментальной последовательности:__


Числовая последовательность $(a_n)$ называется __фундаментальной__ (или последовательностью Коши) , если для любого ε > 0 найдется натуральное число N, такое что для всех m, n ≥ N выполняется неравенство:

$$|a_m - a_n| < ε $$

__Другими словами:__ Члены фундаментальной последовательности "сгущаются" друг к другу по мере увеличения номера члена.


Пример фундаментальной последовательности:


Рассмотрим последовательность $(a_n) = (1/n)$ : 1, 1/2, 1/3, 1/4,...

Эта последовательность является фундаментальной. Для любого ε > 0 можно найти N такое, что для всех m, n ≥ N выполняется:



$$|a_m - a_n| = |\frac{1}{m} - \frac{1}{n}| < ε $$

* Проверка:
* Заметим, что $|a_m - a_n|$ достигает максимального значения при m=N+1 и n=N.
* В этом случае:



$$|a_{N+1} - a_N| = |\frac{1}{N+1} - \frac{1}{N}| < ε $$



* Выбор N: Можно выбрать N таким образом, чтобы это неравенство выполнялось.



### Предел числовой последовательности


__Предел числовой последовательности__ – это значение, к которому стремятся члены последовательности по мере того, как номер члена (индекс) стремится к бесконечности.


__Определение__

Число L называется пределом последовательности $(a_n)$, если для любого сколь угодно малого положительного числа ε > 0 существует такое натуральное число N, что для всех n > N выполняется неравенство:
$|a_n - L| < \varepsilon$.

Важно:

* Не все числовые последовательности имеют предел.
* Если предел существует, он единственный.

__Пример 1__

Последовательность  (1, 1/2, 1/3, 1/4...)  или в краткой записи $\{\frac{1}{n}\}_{n=1}^\infty$ имеет предел 0, так как члены последовательности становятся сколь угодно близкими к 0 при увеличении n.

__Пример 2__

Последовательность Фибоначчи неграничено возрастает и не имеет предела.

__Пример 3__

Последовательность  $\{(-1)^n\}_{n=1}^\infty$ ограничена и не имеет предела.

### Вещественные числа

__Пример__ фундаментальной последовательности рациональных чисел, не имеющей предела в множестве рациональных чисел.

Последовательность $$x_{n+1}=\frac{1}{2}\left(x_n+\frac{2}{x_n}\right).$$

Покажем, что последовательность является фундаментальной.

__Доказательство:__

__1.__  Пусть $x_{n}>1$, тогда $x_{n+1}=\frac{1}{2}\left(x_n+\frac{2}{x_n}\right)>1.$

__2.__ Для любых m > n, имеем:
$$
|x_m - x_n| = \left|\frac{1}{2}\left(x_{m-1} + \frac{2}{x_{m-1}}\right) - \frac{1}{2}\left(x_{n-1} + \frac{2}{x_{n-1}}\right)\right|
$$

$$
=\left|\frac{1}{2}\left(x_{m-1} -x_{n-1}\right) - \frac{1}{2} \left(\frac{2}{x_{n-1}}- \frac{2}{x_{m-1}}\right)\right|
$$

$$
=\left|\frac{1}{2}\left(x_{m-1} -x_{n-1}\right) -\left(\frac{x_{m-1}-x_{n-1}}{x_{n-1}x_{m-1}}\right)\right|
$$

$$
=\left|\frac{1}{2} -\frac{1}{x_{n-1}x_{m-1}}\right|\cdot\left|x_{m-1} -x_{n-1}\right|
$$

__3.__ Из п.1. следует:

$$
\left|\frac{1}{2} -\frac{1}{x_{n-1}x_{m-1}}\right|<\frac{1}{2}.
$$

__4.__ Тогда:  

$$
|x_m - x_n|<\frac{1}{2}|x_{m-1} - x_{n-1}|\, \Rightarrow\, |x_{m+k} - x_{n+k}|<\frac{1}{2^{k}}|x_{m} - x_{n}|.
$$
__5.__ При $k\to\infty,\,\,\ |x_{m+k} - x_{n+k}|\to 0$. То есть, __последовательность фундаментальная.__ 

Решение уравнения
$$x=\frac{1}{2}\left(x+\frac{2}{x}\right)\Rightarrow 2x^2=x^2+2\Rightarrow x=\pm\sqrt{2}.$$

-----

```python
import matplotlib.pyplot as plt
import numpy as np

# Начальное значение последовательности
x0 = 5

# Количество членов последовательности для построения
num_terms = 10

# Массив для хранения значений последовательности
xn = np.zeros(num_terms)
xn[0] = x0

# Вычисляем члены последовательности по рекуррентной формуле
for i in range(1, num_terms):
    xn[i] = (1/2)*(xn[i-1] + 3/(xn[i-1]))

# Создаём график
n = np.arange(num_terms)  
plt.plot(n, xn, marker='o', linestyle='-', color='b')  

plt.xlabel('n')                                      
plt.ylabel('x_n')                              
plt.title('График последовательности x_{n+1} = (1/2)*(x_n + 3/x_n)')            
plt.grid(True)   
plt.show()
```

#### Теорема об иррациональности квадратного корня из двух

__Теорема:__ √2 — иррациональное число.

__Доказательство от противного:__

__1. Предположение:__ Предположим, что √2 рационально. Это означает, что его можно представить в виде дроби $\frac{a}{b}$, где a и b — целые числа без общих делителей (кроме 1), а b ≠ 0.

__2. Возведение в квадрат:__
$$ (\sqrt{2})^2 = \left( \frac{a}{b} \right)^2 $$
$$ 2 = \frac{a^2}{b^2}$$

__3. Умножение на b²:__

$$ 2b^2 = a^2$$

__4. Четность a:__ Это уравнение означает, что $a^2$ — четное число. Следовательно, a само по себе также должно быть четным.

__5. Представление a:__ Так как a — четное число, его можно записать в виде:
$$ a = 2k $$ , где k — целое число.

__6. Подстановка:__
$$ 2b^2 = (2k)^2$$
$$ 2b^2=4k^2$$

__7. Четность b:__
$$ b^2 = 2k^2 $$
Следовательно, b должно быть четным.

__8. Противорчие:__ Мы пришли к выводу, что как a, так и b должны быть четными числами. Но это противоречит нашему первоначальному предположению, что $\frac{a}{b}$ — дробь в простейшей форме без общих делителей.

__9. Заключение:__
Поскольку наше предположение привело к противоречию, то оно ложно. Следовательно, √2 не может быть рациональным числом и является иррациональным.

#### Дедекиндово сечение

* Представим множество рациональных чисел Q как упорядоченное по величине множество.
* Дедекиндово сечение на множестве Q - это разбиение его на два непустых подмножества A и B, удовлетворяющее следующим условиям:

__1.__ A ∪ B = Q: Объединение подмножеств A и B равно всему множеству рациональных чисел Q.

__2.__ a < b для всех a ∈ A и b ∈ B: Все элементы в A меньше всех элементов в B.

__3.__ A не содержит наибольшего элемента: Не существует максимального элемента в подмножестве A.



__Теорема:__ Объединение множества рациональных чисел Q и множества всех дедекиндовых сечений D образует полное упорядоченное множество.


__Доказательство:__


__1.__ Определение порядка: Определим порядок на объединении Q∪D следующим образом:
Для любых x, y ∈ Q*, порядок наследуется из обычного порядка рациональных чисел.
Для любого дедекиндова сечения α и рационального числа r*:
α < r, если существует рациональное число q ∈ α такое, что q < r*.

__2.__ Полнота Q∪D следует из того, как мы определили дедекиндовы сечения:

* Любое ограниченное сверху множество в Q∪D имеет точную верхнюю грань (супремум), которая либо является рациональным числом, либо представлена дедекиндовым сечением. Это свойство вытекает из аксиомы полноты вещественных чисел, на которой основано понятие дедекиндова сечения.



__Интерпретация:__

* Дедекиндово сечение можно рассматривать как "разрыв" на числовой прямой Q, который делит рациональные числа на две части: A (элементы меньше разрыва) и B (элементы больше разрыва).
* Вещественное число: Каждое дедекиндово сечение определяет одно вещественное число.

__Построение R:__

1. Множество всех дедекиндовых сечений Q образует множество вещественных чисел R.

2. Рациональные числа в R: Рациональное число r соответствует такому дедекиндову сечению, где A содержит все рациональные числа меньше r, а B - все рациональные числа больше r.
3. Иррациональные числа в R соответствуют таким дедекиндовым сечениям, которые не могут быть представлены одним рациональным числом.

__Пример:__

Рассмотрим иррациональное число √2. Его можно представить как дедекиндово сечение Q следующим образом:

* A = {x ∈ Q | x² < 2} (все рациональные числа, квадрат которых меньше 2)
* B = {x ∈ Q | x² > 2} (все рациональные числа, квадрат которых больше 2)


__Преимущества этого подхода:__



* Строгость: Построение на основе дедекиндовых сечений строго математически и не зависит от геометрической интуиции.
* Обоснование всех вещественных чисел: Оно позволяет формально определить все вещественные числа, включая как рациональные, так и иррациональные.
* Свойства: Определение дедекиндовых сечений естественным образом приводит к важным свойствам вещественных чисел, таким как непрерывность.

Важно отметить: Построение R с помощью дедекиндовых сечений - это один из способов формализовать понятие вещественного числа. Существует и другой подход - через классы эквивалентности последовательностей Коши рациональных чисел.

#### Теорема о несчетности множества вещественных чисел

__Теорема (Кантор):__ Множество всех вещественных чисел (R) несчётно.



Это означает, что не существует взаимно однозначного соответствия между множеством натуральных чисел (N) и множеством вещественных чисел. Другими словами, вещественные числа нельзя пронумеровать.

Более формально:
Не существует функции f: N → R, которая является биекцией (взаимно однозначным отображением).

__Доказательство теоремы Кантора о несчётности вещественных чисел__


__Доказательство от противного:__


__1. Предположение:__ Предположим, что множество всех вещественных чисел (R) счетно. Это означает, что существует биекция f: N → R, где N — множество натуральных чисел.
* Другими словами, мы можем пронумеровать все вещественные числа.

__2. Представление вещественных чисел:__ Представим каждое вещественное число x в интервале [0,1] в виде бесконечной десятичной дроби и построим таблицу из этих чисел:

0.__1__ 238023849...\
0.2 __3__ 28948302...\
0.84 __9__ 3049584...\
0.172 __6__ 354049...\
0.2034 __9__ 56094...\
0.92874 __6__ 5271...\
0.645342 __3__ 152...\
0.8574635 __2__ 41...\
0.00100100 __1__ 0...\
0.111191919 __1__ ...




__3. Построение нового числа:__ Построим новое вещественное число *Y* следующим образом:
* Рассмотрим n-ю цифру после запятой n-го вещественного числа в нашем предполагаемом списке всех вещественных чисел (соответствующего f(n)).

* Если эта цифра равна k, то k-я цифра нашего нового числа *Y* будет "k+1" (или 0, если k = 9).
* Для приведенной таблицы: *Y*=__0.2407074322__ ...

__4. Противоречие:__
* Число *Y* отличается от всех чисел в списке f по крайней мере на одной цифре (n-й).

* Это означает, что наше новое число *Y* не входит в список f(1), f(2), f(3)...


__5. Вывод:__ Мы получили противоречие с предположением о том, что f является биекцией между N и R. Следовательно, множество вещественных чисел несчетно.

#### Теорема Лиувилля об аппроксимации иррациональных чисел рациональными



__Формулировка:__
Для любого алгебраического числа α (корня многочлена с целыми коэффициентами) существует константа C > 0, такая что для любых рациональных чисел p/q (где q ≠ 0):

$$|\alpha - \frac{p}{q}| ≥ \frac{C}{q^d}$$



где d — степень многочлена наименьшей степени с целыми коэффициентами, корнем которого является α.



__Простыми словами:__ Теорема Лиувилля утверждает, что иррациональные числа, которые являются корнями полиномов с целыми коэффициентами (алгебраические), могут быть аппроксимированы рациональными числами только с ограниченной точностью.

__Ключевые моменты:__

* Аппроксимация: Теорема касается того, насколько близко мы можем приблизиться к иррациональному числу α с помощью рационального числа p/q.
* Алгебраические числа: Теорема применима только к алгебраическим числам (корням многочленов с целыми коэффициентами), а не ко всем иррациональным числам.

## Теоремы о пределах числовых последовательностей

### Теорема о единственности предела числовой последовательности:


Если числовая последовательность $(a_n)$ имеет предел, то этот предел единственный.


__Доказательство:__

Предположим противное: что последовательность $(a_n)$ имеет два разных предела - L и M, где L ≠ M.

По определению предела для ε = |L-M|/2 > 0 существуют такие натуральные числа N₁ и N₂, что:
* Для всех n > N₁, выполняется $|a_n - L| < \varepsilon$.
* Для всех n > N₂, выполняется $|a_n - M| < \varepsilon$.

Выберем n₀ = max(N₁, N₂). Тогда для всех n > n₀ выполняются оба неравенства:

* $|a_n - L| < |L-M|/2$
* $|a_n - M| < |L-M|/2$

Сложив эти неравенства, получим:

$|a_n - L| + |a_n - M| < |L-M|$.

По неравенству треугольника для модуля:
$|L - M| ≤ |a_n - L| + |a_n - M|$

Отсюда получаем:

$|L - M| < |L-M|$,

что является противоречием.

Следовательно, наше предположение о существовании двух разных пределов неверно, и предел числовой последовательности единственен.

-----------------

### Теорема о пределе монотонной последовательности 


__Теорема__

Каждая монотонно растущая ограниченная сверху последовательность имеет предел.

__Доказательство:__

1. Ограниченность: Пусть $(a_n)$ — монотонно растущая и ограниченная сверху последовательность, т.е. существует число M такое, что $a_n \le M$ для всех n.

2. Множество верхних границ: Обозначим через S множество всех верхних граней последовательности $(a_n)$. По определению, это множество непусто (т.к. оно содержит M), и S ограничено снизу элементами самой последовательности: $a_n \in S$ для всех n.

3. Наименьшая верхняя грань: Согласно свойству полноты вещественных чисел, существует наименьшая верхняя грань (supremum) множества S, которую обозначим через L:

L = sup S


4. Доказательство предела: Покажем, что L является пределом последовательности $(a_n)$:
Для любого ε > 0, число L - ε не является верхней гранью множества S, так как L — наименьшая верхняя грань. Значит, существует член последовательности $a_{N}$ такой, что $a_{N} > L - ε$.

* По монотонности последовательности $(a_n)$, для всех n ≥ N выполняется:
$a_n \ge a_{N} > L - ε$.

Так как L является верхней гранью множества S, то $a_n \le L$ для всех n.

Следовательно, для всех n ≥ N выполняется неравенство:


$$L - ε < a_n ≤ L$$

* Отсюда получаем:

$$|a_n - L| < ε$$

для всех n ≥ N. Это и означает по определению, что L является пределом последовательности $(a_n)$. 

-----

###  Подпоследовательность и частичные пределы

__Определение:__

Подпоследовательность — это последовательность, полученная из исходной последовательности путем отбора некоторых ее членов, сохраняя их порядок следования.


__Более формально:__

Пусть $(a_n)$ - исходная числовая последовательность. Подпоследовательность $(a{n_k})$ определяется бесконечной возрастающей последовательностью индексов $(n_k)$, где каждый $n_k$ — натуральное число:

$$1 \le n_1 < n_2 < n_3 < ... $$

Тогда подпоследовательность записывается как:

$$ (a_{n_1}, a_{n_2}, a_{n_3}, ...) $$


__Примеры:__

* $(a_{2k})$ — подпоследовательность, состоящая из членов с четными номерами.
* $(a_{3k+1})$ - подпоследовательность, состоящая из членов номерами, у которых остаток от деления на 3 равен 1.

__Важно:__


* Подпоследовательность всегда наследует порядок следования элементов от исходной последовательности.
* Каждая исходная последовательность является подпоследовательностью самой себя (с индексами $(n_k) = (1,2,3,...)$.

-------------

### Верхний  и нижний пределы 


__Определение__

Верхним (нижним) пределом числовой последовательности $\{a_n\}$ называется наибольший (наименьший) из всех частичных пределов этой последовательности.

Обозначения:

* $$\overline{\lim_{n\to\infty}} {a_n}$$

* $$\underline{\lim}_{n\to\infty} {a_n}$$


__Определение более формально:__

Число L называется верхним пределом последовательности $(a_n)$, если:
1. L является верхней гранью множества $\{a_n\}$, т.е. $a_n \le L$ для всех n∈ℕ.
2. Для любого ε > 0 существует элемент aₙ в последовательности, такой что L - ε < aₙ.

__Важно:__


* Верхний предел может быть конечным числом или +∞.
* Если последовательность ограничена сверху, то ее верхний предел совпадает с supremum множества всех ее членов.

### Теорема Больцано-Вейерштрасса

Каждая ограниченная последовательность вещественных чисел содержит сходящуюся подпоследовательность.


Другими словами: Если все члены числовой последовательности находятся в пределах некоторого конечного интервала, то можно выделить из этой последовательности бесконечную подпоследовательность, которая сходится к определенному пределу.


__Доказательство теоремы Больцано-Вейерштрасса:__


1. Ограниченность: Пусть $(a_n)$ — ограниченная последовательность вещественных чисел.
* Это означает, что существует число M > 0 такое, что для всех n выполняется $|a_n| \leq M$.


2. Деление интервала пополам:

* Рассмотрим замкнутый интервал [−M, M], который содержит все члены последовательности $(a_n)$.
* Разделим этот интервал на два равных подинтервала: [-M, 0] и [0, M].

3. Выбор подинтервала:

* Поскольку в $(a_n)$ бесконечное число членов, по крайней мере один из этих двух подынтервалов должен содержать бесконечно много членов последовательности $(a_n)$.
* Обозначим этот подынтервал как $[b_1, c_1]$.

4. Рекурсия:

* Разделим интервал $[b_1, c_1]$ на два равных подинтервала: $[b_1, (b_1+c_1)/2]$ и $[(b_1+c_1)/2, c_1]$.
* Опять же, один из этих двух подынтервалов должен содержать бесконечно много членов последовательности. Обозначим его как $[b_2, c_2]$.

5. Повторение:

* Продолжаем этот процесс рекурсивно: делим выбранный интервал пополам и выбираем тот подынтервал, который содержит бесконечно много членов последовательности.
* Это создает последовательность вложенных отрезков $[b_1, c_1]$ ⊇ $[b_2, c_2]$ ⊇ $[b_3, c_3]$ ⊇ ...

6. Свойства вложенных отрезков:


* Длина стремится к нулю: Длина каждого интервала $[b_n, c_n]$ равна $(c_n - b_n)/2^n$. По построению, эта длина стремится к 0 при n → ∞.

* Вложенность: Каждый последующий отрезок содержится в предыдущем: $[b_1, c_1]$ ⊇ $[b_2, c_2]$ ⊇ ...
* Непустота: Все отрезки не пусты, так как каждый из них содержит бесконечно много членов последовательности $(a_n)$.

7. Применение теоремы о сходящейся вложенной последовательности интервалов:
* По теореме о сходящейся вложенной последовательности интервалов (которая вытекает из аксиомы полноты вещественных чисел), существует единственное число L, которое принадлежит всем отрезкам $[b_n, c_n]$.

8. Подпоследовательность:
* Для каждого n мы можем выбрать член $a_{k_n}$ последовательности $(a_n)$, который принадлежит интервалу $[b_n, c_n]$.
* По построению, эта подпоследовательность $(a_{k_1}, a_{k_2}, ..., a_{k_n}, ...)$ сходится к L.

9. Вывод: Мы доказали, что из ограниченной последовательности $(a_n) $ можно выделить сходящуюся подпоследовательность.


----------

```python
import matplotlib.pyplot as plt
import numpy as np

# Иллюстрация для первых трех шагов рекурсии в доказательстве

import matplotlib.pyplot as plt
import numpy as np

# Генерируем значения n для последовательностей
n1 = np.arange(1, 50, 1)  
n2 = np.arange(1, 50, 1)
n3 = np.arange(1, 50, 1)

# Вычисляем члены подпоследовательностей
y1 = 1/3+1/n1*np.sin(n1)           
y2 = 1/3*np.sin(n2+1/n2)
y3 = 1/3+1/2+1/4*np.sin(2*n3+1/n3)

# Находим верхние и нижние гранипоследовательности
upper_bound = np.ones_like(n) 
lower_bound = np.zeros_like(n)
upper_bound = 1/3+1/2+1/4 # Верхняя грань
lower_bound = -1/3 # Нижняя грань
average_line=(upper_bound+lower_bound)/2
second_line=(average_line+lower_bound)/2
third_line=(second_line+average_line)/2

# Рисуем последовательность
plt.plot(n1, y1, marker='o', linestyle='', color='b')  
plt.plot(n2, y2, marker='o', linestyle='', color='b')
plt.plot(n3, y3, marker='o', linestyle='', color='b')

# Рисуем разбиение на области 
plt.plot(n, np.full(len(n), upper_bound), linestyle='-', color='k') 
plt.plot(n, np.full(len(n), lower_bound), linestyle='-', color='k')
plt.plot(n, np.full(len(n), average_line), linestyle='-', color='r')
plt.plot(n, np.full(len(n), second_line), linestyle='-', color='b')
plt.plot(n, np.full(len(n), third_line), linestyle='-', color='g')

plt.title("Иллюстрация теоремы Больцано-Вейерштрасса")
plt.show()
```

## Канторово множество


__Определение:__
Канторово множество — это подмножество вещественных чисел на интервале [0,1], которое получается из отрезка [0,1] путем последовательного удаления "средних третьих" каждого оставшегося отрезка.


__Построение:__

1. Удаляем открытый интервал (1/3; 2/3) из [0,1].
2. Из оставшихся двух отрезков [0, 1/3] и [2/3, 1], удаляем их средние трети: (1/9; 2/9) и (7/9; 8/9).

3. Продолжаем этот процесс бесконечно: на каждом шаге из оставшихся отрезков удаляем среднюю треть.
4. Канторово множество — это все точки, которые остаются после бесконечного числа таких удалений.
___


#### Тернарное представление

Множество Кантора тесно связано с тернарной системой счисления (системой счисления по основанию 3). Вот как элементы множества Кантора можно представить в этой системе:

1. Тернарное представление: Каждое число в интервале [0, 1] может быть представлено в виде бесконечной дробной части в тернарной системе счисления. Например:
* 2/3 = 0.2 

* 1/2=0.11111(1) Действительно:

```python
1/3+1/3**2+1/3**3+1/3**4+1/3**5+1/3**7+1/3**8
```

2. Удаление средних третей: При построении множества Кантора мы удаляем среднюю треть каждого отрезка на каждом шаге. В тернарной системе счисления это соответствует удалению чисел, которые содержат цифру "1" в своей дробной части.

3. Элементы Канторова множества: Элементы множества Кантора — это те числа в интервале [0, 1], которые могут быть представлены в тернарной системе счисления без использования цифры "1".

Примеры:


* Число 1/3 = 0.1 (в тернарном представлении) принадлежит множеству Кантора. Его можно представить в виде:1/3=0.0222222(2)


```python
2/3**2+2/3**3+2/3**4+2/3**5+2/3**6+2/3**7+2/3**8
```




* Число 2/9 = 0.02 (в тернарном представлении) также принадлежит множеству Кантора. 

Вывод:


Множество Кантора можно представить как множество всех чисел в интервале [0,1], которые имеют тернарное представление без цифры "1".

#### Длина Канторова множества

Доказательство того, что длина Канторова множества равна нулю



Для доказательства, что длина (или мера Лебега) Канторова множества C равна нулю, мы можем использовать следующую стратегию:

1. Построение покрытия интервалами:

* На каждом шаге построения Канторова множества мы удаляем среднюю треть каждого оставшегося отрезка.
* Обозначим длину удаляемых отрезков на n-ом шаге как Lₙ (например, L₁ = 1/3, L₂ = 2/9, L₃ = 4/27 и т.д.).

2. Суммирование длин удаленных интервалов:


* Общая длина всех удаленных интервалов равна сумме L₁, L₂, L₃,...
* Эта сумма является геометрическим рядом: (1/3) + (2/9) + (4/27) + ...

3. Сходимость геометрического ряда:

* Геометрический ряд слагается из членов вида a ⋅ rⁿ, где a = 1/3 и r = 2/3 (знаменатель).
* Поскольку |r| < 1, этот ряд сходится к сумме a/(1-r) = (1/3)/(1-(2/3)) = 1.

4. Вывод:


* Мы удалили интервалы общей длины 1 (длина исходного отрезка [0,1]).
* Следовательно, длина оставшегося Канторова множества равна 1 - 1 = 0.
